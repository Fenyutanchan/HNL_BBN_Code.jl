## Scope

- Number density evolution $n_N(a) \approx n_N^{(\mathrm{fi/fo})}(a) e^{-\Gamma_N t}$ (Eq. 2.14)
- Freeze-in abundance $n_N^{(\mathrm{fi})}$ (Eq. 2.15) and freeze-out $n_N^{(\mathrm{fo})}$ (Eq. 2.16)
- $R_\rho \equiv (\rho_N/\rho_{\nu_\alpha})_{T=1\,\mathrm{MeV}}$ and $R_n \equiv (n_N/n_{\nu_\alpha})_{T=1\,\mathrm{MeV}}$ (Eq. 2.17)
- Total decay width $\Gamma_N$ using $\Gamma_{N\to 3\nu}$ and branching ratios
- $(m_N, U^2)$ parameter scan: $m_N \in [0.05, 10]$ GeV, $U_e^2 \in [10^{-18}, 10^{-4}]$
- Thermalization boundary identification
